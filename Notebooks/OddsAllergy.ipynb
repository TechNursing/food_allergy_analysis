```python
import pandas as pd
import japanize_matplotlib

# 人種の影響が出ない様にそれぞれの人種数を合わせたデータを使用する
allergy_data = pd.read_csv('../Data/race_sample.csv')
display(allergy_data.head(10))
allergy_data.dtypes
```

```python
# オッズ比を見たいアトピーと喘息、食物アレルギーのカラムのみにする
allergy_data = allergy_data.reindex(columns=['ATOPIC_DERM_START', 'ASTHMA_START',
                                          'SHELLFISH_ALG_START', 'FISH_ALG_START','MILK_ALG_START',
                                          'SOY_ALG_START' ,'EGG_ALG_START', 'WHEAT_ALG_START',
                                          'PEANUT_ALG_START', 'SESAME_ALG_START','TREENUT_ALG_START',
                                          'WALNUT_ALG_START' ,'PECAN_ALG_START', 'PISTACH_ALG_START',
                                          'ALMOND_ALG_START', 'BRAZIL_ALG_START','HAZELNUT_ALG_START',
                                          'CASHEW_ALG_START'])
```

```python
# アトピー性皮膚炎の患児で食物アレルギー感作の有無で喘息の発症率を見る
```

```python
# クロス集計とオッズ比
```

```python
# アレルギー感作があるアトピー性皮膚炎、アレルギー感作がないアトピー性皮膚炎、喘息発症あり、喘息発症なしのクロス表を作成する。
# 食物アレルギー感作を保持の有無のみにする
allergy_data['SHELLFISH_ALG_START'] = allergy_data['SHELLFISH_ALG_START'].notna()
allergy_data['FISH_ALG_START'] = allergy_data['FISH_ALG_START'].notna()
allergy_data['MILK_ALG_START'] = allergy_data['MILK_ALG_START'].notna()
allergy_data['SOY_ALG_START'] = allergy_data['SOY_ALG_START'].notna()
allergy_data['EGG_ALG_START'] = allergy_data['EGG_ALG_START'].notna()
allergy_data['WHEAT_ALG_START'] = allergy_data['WHEAT_ALG_START'].notna()
allergy_data['PEANUT_ALG_START'] = allergy_data['PEANUT_ALG_START'].notna()
allergy_data['SESAME_ALG_START'] = allergy_data['SESAME_ALG_START'].notna()
allergy_data['TREENUT_ALG_START'] = allergy_data['TREENUT_ALG_START'].notna()
allergy_data['WALNUT_ALG_START'] = allergy_data['WALNUT_ALG_START'].notna()
allergy_data['PECAN_ALG_START'] = allergy_data['PECAN_ALG_START'].notna()
allergy_data['PISTACH_ALG_START'] = allergy_data['PISTACH_ALG_START'].notna()
allergy_data['ALMOND_ALG_START'] = allergy_data['ALMOND_ALG_START'].notna()
allergy_data['BRAZIL_ALG_START'] = allergy_data['BRAZIL_ALG_START'].notna()
allergy_data['HAZELNUT_ALG_START'] = allergy_data['HAZELNUT_ALG_START'].notna()
allergy_data['CASHEW_ALG_START'] = allergy_data['CASHEW_ALG_START'].notna()

# アレルギー疾患を保持の有無のみにする
allergy_data['ATOPIC_DERM_START'] = allergy_data['ATOPIC_DERM_START'].notna()
allergy_data['ASTHMA_START'] = allergy_data['ASTHMA_START'].notna()
```

```python
# 食物アレルギー感作があるアトピー性皮膚炎の喘息あり
atopic_positive = allergy_data[allergy_data['ATOPIC_DERM_START'] == True]
allergen_positive = atopic_positive[(atopic_positive['SHELLFISH_ALG_START'] == True)|
                                    (atopic_positive['FISH_ALG_START'] == True)|
                                    (atopic_positive['MILK_ALG_START'] == True)|
                                    (atopic_positive['SOY_ALG_START'] == True)|
                                    (atopic_positive['EGG_ALG_START'] == True)|
                                    (atopic_positive['WHEAT_ALG_START'] == True)|
                                    (atopic_positive['PEANUT_ALG_START'] == True)|
                                    (atopic_positive['SESAME_ALG_START'] == True)|
                                    (atopic_positive['TREENUT_ALG_START'] == True)|
                                    (atopic_positive['WALNUT_ALG_START'] == True)|
                                    (atopic_positive['PECAN_ALG_START'] == True)|
                                    (atopic_positive['PISTACH_ALG_START'] == True)|
                                    (atopic_positive['ALMOND_ALG_START'] == True)|
                                    (atopic_positive['BRAZIL_ALG_START'] == True)|
                                    (atopic_positive['HAZELNUT_ALG_START'] == True)|
                                    (atopic_positive['CASHEW_ALG_START'] == True)
                                   ]
allergen_positive_asth = allergen_positive[allergen_positive['ASTHMA_START'] == True]
allergen_positive_asth_number = len(allergen_positive_asth)
print(allergen_positive_asth_number)
```

```python
# 食物アレルギー感作があるアトピー性皮膚炎の喘息なし
allergen_positive_only = allergen_positive[allergen_positive['ASTHMA_START'] == False]
allergen_positive_only_number = len(allergen_positive_only)
print(allergen_positive_only_number)
```

```python
# 食物アレルギー感作がないアトピー性皮膚炎の喘息あり
allergen_negative = atopic_positive[(atopic_positive['SHELLFISH_ALG_START'] == False)&
                                    (atopic_positive['FISH_ALG_START'] == False)&
                                    (atopic_positive['MILK_ALG_START'] == False)&
                                    (atopic_positive['SOY_ALG_START'] == False)&
                                    (atopic_positive['EGG_ALG_START'] == False)&
                                    (atopic_positive['WHEAT_ALG_START'] == False)&
                                    (atopic_positive['PEANUT_ALG_START'] == False)&
                                    (atopic_positive['SESAME_ALG_START'] == False)&
                                    (atopic_positive['TREENUT_ALG_START'] == False)&
                                    (atopic_positive['WALNUT_ALG_START'] == False)&
                                    (atopic_positive['PECAN_ALG_START'] == False)&
                                    (atopic_positive['PISTACH_ALG_START'] == False)&
                                    (atopic_positive['ALMOND_ALG_START'] == False)&
                                    (atopic_positive['BRAZIL_ALG_START'] == False)&
                                    (atopic_positive['HAZELNUT_ALG_START'] == False)&
                                    (atopic_positive['CASHEW_ALG_START'] == False)
                                   ]
allergen_negative_asth = allergen_negative[allergen_negative['ASTHMA_START'] == True]
allergen_negative_asth_number = len(allergen_negative_asth)
print(allergen_negative_asth_number)
```

```python
# 食物アレルギー感作がないアトピー性皮膚炎の喘息なし
atopic_only = allergen_negative[allergen_negative['ASTHMA_START'] == False]
atopic_only_number = len(atopic_only)
print(atopic_only_number)
```

```python
# テーブルでクロス表作成
col = ['喘息発症あり', '喘息発症なし', '合計']
ind = ['アレルギー感作があるアトピー性皮膚炎', 'アレルギー感作がないアトピー性皮膚炎', '合計']
sum1 = sum([allergen_positive_asth_number, allergen_positive_only_number])
sum2 = sum([allergen_negative_asth_number, atopic_only_number])
sum3 = sum([allergen_positive_asth_number, allergen_negative_asth_number])
sum4 = sum([allergen_positive_only_number, atopic_only_number])
sum5 = sum([sum1, sum2])
list = [
    [allergen_positive_asth_number, allergen_positive_only_number, sum1],
    [allergen_negative_asth_number, atopic_only_number, sum2], 
    [sum3,sum4,sum5]
]

print("アトピー性皮膚炎患児の総数\n", len(atopic_positive))
atop_asth_allergy_table = pd.DataFrame(list, index=ind, columns=col)
display(atop_asth_allergy_table)
```

```python
# オッズ比とP値を求める
from scipy.stats import fisher_exact

# オッズ用のテーブル作成
col = ['喘息発症あり', '喘息発症なし']
ind = ['アレルギー感作があるアトピー性皮膚炎', 'アレルギー感作がないアトピー性皮膚炎']
list2 = [
    [allergen_positive_asth_number, allergen_positive_only_number],
    [allergen_negative_asth_number, atopic_only_number]
]
odds_table = pd.DataFrame(list2, index=ind, columns=col)
display(odds_table)

# フィッシャーの正確確率検定を用いて、テーブルからオッズ比とP値を求める
odds, p = fisher_exact(odds_table)
print('アレルギー感作があるアトピー性皮膚炎とアレルギー感作がないアトピー性皮膚炎の喘息発症率の差\n', "オッズ比: ", odds, "\n p-value: ", p)
```

```python
print("＊補足 p-value: ", p, "= 0.0000000000000000000000000019451990747405738")
print('アレルギー感作があるアトピー性皮膚炎が喘息を発症する確率は、アレルギー感作がないアトピー性皮膚炎の場合に比べてオッズ比が\n',
odds,'と高い事が分かった。また、P値も', p, 'と信頼性が非常に高い結果となった。')
```
